```python
import json
import matplotlib.pyplot as plit

with open ("land.json", encoding="utf-8") as fil:
    data = json.load(fil)

data[0]
```

```python
verdensdeler = {}
for land in data:
    print(land['verdensdel'])
    if land['verdensdel'] not in verdensdeler:
        print("fant en ny verdensdel! putter den i ordboka")
        verdensdeler[land['verdensdel']] = 1
    else:
        verdensdeler[land['verdensdel']]+= 1

print(verdensdeler)
```

```python
land = list(verdensdeler.keys())#gjør nøklene til liste
antall = list(verdensdeler.values())# gjør verdiene til liste

#oppgave plott land og antall
plt.bar(land,antall)
plt.title("antall land på topplisten")
plt.xlabel("verdensdeler")
plt.ylabel("antall")
```
